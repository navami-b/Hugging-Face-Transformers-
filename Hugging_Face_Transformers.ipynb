```python
# Cell 2 — run a small text2text model (google/flan-t5-small)
from transformers import pipeline

model_name = "google/flan-t5-small"   # small-ish, suitable for Colab CPU
pipe = pipeline("text2text-generation", model=model_name, device_map=None)  # CPU by default

prompt = "Explain how rainbows are formed"
out = pipe(prompt, max_new_tokens=200, do_sample=False)  # deterministic
print("Prompt:", prompt)
print("\nAnswer:\n", out[0]["generated_text"].strip())
```
